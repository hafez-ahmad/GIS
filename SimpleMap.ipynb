```python
from pydap.client import open_url
```

```python
URL = 'http://nomads.ncep.noaa.gov:9090/dods/rtofs/rtofs_global20141008/rtofs_glo_3dz_forecast_daily_temp'
dataset = open_url(URL)

print type(dataset)
```

```python
print dataset.keys()
```

```python
temp = dataset['temperature']

print type(temp)
```

```python
print temp.dimensions
```

```python
print temp.shape
```

```python
lat = temp['lat'][1700:2100]
lon = temp['lon'][3000:3300]


minlon = min(lon)
maxlon = max(lon)
minlat = min(lat)
maxlat = max(lat)
```

```python
import numpy as np

lon = np.where(np.greater_equal(lon,500),np.nan,lon)
```

```python
dataOriginal = temp['temperature'][1,1,1700:2100,3000:3300]
# dataOriginal = temp['temperature'][1,1,:,:]
```

```python
datashape = dataOriginal.shape

data = dataOriginal.reshape(datashape[2],datashape[3])

data = np.ma.masked_where(data>500,data)


from mpl_toolkits.basemap import shiftgrid
dat, lon = shiftgrid(330., data, lon, start=False)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.pcolormesh(data)

# print data
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
%matplotlib inline

# Initialize projection and map limits/properties
m = Basemap(projection='mill',llcrnrlon=minlon,urcrnrlon=maxlon,llcrnrlat=minlat,urcrnrlat=maxlat,resolution='l')

x, y = m(lonMat, latMat)

cs = m.pcolormesh(lonMat, latMat, data, cmap=plt.cm.jet)

# Draw package-included stuff
m.drawcoastlines(linewidth=0.25)

plt.show()
```
