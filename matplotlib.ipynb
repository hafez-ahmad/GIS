```python
x=[1,2,3,4,23,44]
y=[2,3,4,5,14,66]
x1=[1,4,6,18,23]
y1=[2,3,7,14,20]
x2=[1,4,8,20,55]
y2=[2,3,10,14,50]
plt.title('General plot',fontsize='19')
plt.style.use('ggplot')
plt.xlabel('y axis')
plt.ylabel('x axis')

plt.plot(x,y,label='one')
plt.plot(x1,y1, label='two')
plt.plot(x2,y2,label='three')
plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
#pie chart
x=[5,6,3,4,17,20,8]
y=['working','sleeping','eating','walking','python','notebook','ipython']
colors=['k','m','y','r','g','b','c']
plt.pie(x,labels=y,colors=colors,startangle=90,shadow=True,autopct='%1.1f%%')
plt.legend()
plt.title('Pie chart on my daily life',fontsize='29')
plt.xlabel('y axis')
#plt.ylabel('x axis')
plt.show()
```
